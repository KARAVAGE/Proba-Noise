#**Noise Bernoulli**

```python
from google.colab import drive

# Mount Google Drive
# drive.mount('/content/drive')

import numpy as np
import cv2
import matplotlib.pyplot as plt

# Path to the image in your Google Drive
image_path = 'image.jpg'

# Load the image
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
print("Original image shape:", image.shape)

# Function to add Bernoulli noise to the image
def add_bernoulli_noise(image, probability):
    # Generate random noise based on Bernoulli distribution
    noise = np.random.binomial(1, probability, size=image.shape).astype(np.uint8) * 255
    # Apply the noise to the image
    noisy_image = cv2.add(image, noise)
    return noisy_image

# Function to remove noise from the image
def remove_noise(image, probability):
    # Calculate the amount of noise to remove
    noise_to_remove = int(probability * 255)
    # Thresholding to remove noise
    _, thresholded_image = cv2.threshold(image, noise_to_remove, 255, cv2.THRESH_BINARY)
    return thresholded_image

# Start probability for adding noise
add_probability = 0.05

# Loop until probability reaches 1 for adding noise
while add_probability <= 1:
    # Add Bernoulli noise to the image
    noisy_image = add_bernoulli_noise(image, add_probability)

    # Display the noisy image
    plt.figure(figsize=(10, 5))
    plt.imshow(noisy_image, cmap='gray')
    plt.title(f'Noisy Image (Add Probability: {add_probability})')
    plt.axis('off')
    plt.show()

    # Increment probability by 0.05 for adding noise
    add_probability += 0.05

# Start probability for removing noise
remove_probability = 1

# Loop until probability reaches 0 for removing noise
while remove_probability >= 0:
    # Remove noise from the image
    restored_image = remove_noise(noisy_image, remove_probability)

    # Display the restored image
    plt.figure(figsize=(10, 5))
    plt.imshow(restored_image, cmap='gray')
    plt.title(f'Restored Image (Remove Probability: {remove_probability})')
    plt.axis('off')
    plt.show()

    # Decrement probability by 0.05 for removing noise
    remove_probability -= 0.05
```

# **Noise Binomial**

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Path to the image in your Google Drive
image_path = 'image.jpg'

# Load the image
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
print("Original image shape:", image.shape)

# Function to add Binomial noise to the image
def add_binomial_noise(image, probability):
    # Generate random noise based on Binomial distribution
    noise = np.random.binomial(255, probability, size=image.shape).astype(np.uint8)
    # Apply the noise to the image
    noisy_image = cv2.add(image, noise)
    return noisy_image

# Function to remove noise from the image
def remove_noise(image, probability):
    # Calculate the threshold value to remove noise
    threshold_value = int((1 - probability) * 255)
    # Thresholding to remove noise
    _, thresholded_image = cv2.threshold(image, threshold_value, 255, cv2.THRESH_BINARY)
    return thresholded_image

# Start probability for adding noise
add_probability = 0.05

# Loop until probability reaches 1 for adding noise
while add_probability <= 1:
    # Add Binomial noise to the image
    noisy_image = add_binomial_noise(image, add_probability)

    # Display the noisy image
    plt.figure(figsize=(10, 5))
    plt.imshow(noisy_image, cmap='gray')
    plt.title(f'Noisy Image (Add Probability: {add_probability})')
    plt.axis('off')
    plt.show()

    # Increment probability by 0.05 for adding noise
    add_probability += 0.05

# Start probability for removing noise
remove_probability = 1

# Loop until probability reaches 0 for removing noise
while remove_probability >= 0:
    # Remove noise from the image
    restored_image = remove_noise(noisy_image, remove_probability)

    # Display the restored image
    plt.figure(figsize=(10, 5))
    plt.imshow(restored_image, cmap='gray')
    plt.title(f'Restored Image (Remove Probability: {remove_probability})')
    plt.axis('off')
    plt.show()

    # Decrement probability by 0.05 for removing noise
    remove_probability -= 0.05
```

# **Noise Expo**

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Path to the image in your Google Drive
image_path = 'image.jpg'

# Load the image
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
print("Original image shape:", image.shape)

# Function to add Exponential noise to the image
def add_exponential_noise(image, scale):
    # Generate random noise based on Exponential distribution
    noise = np.random.exponential(scale, size=image.shape).astype(np.uint8)
    # Apply the noise to the image
    noisy_image = cv2.add(image, noise)
    return noisy_image

# Function to remove noise from the image
def remove_noise(image, scale):
    # Estimate the noise based on the Exponential distribution
    estimated_noise = np.random.exponential(scale, size=image.shape).astype(np.uint8)
    # Remove the estimated noise from the image
    denoised_image = cv2.subtract(image, estimated_noise)
    return denoised_image

# Start scale for adding noise
add_scale = 50

# Loop until scale reaches 100 for adding noise
while add_scale <= 100:
    # Add Exponential noise to the image
    noisy_image = add_exponential_noise(image, add_scale)

    # Display the noisy image
    plt.figure(figsize=(10, 5))
    plt.imshow(noisy_image, cmap='gray')
    plt.title(f'Noisy Image (Add Scale: {add_scale})')
    plt.axis('off')
    plt.show()

    # Increment scale by 10 for adding noise
    add_scale += 10

# Start scale for removing noise
remove_scale = 100

# Loop until scale reaches 0 for removing noise
while remove_scale >= 0:
    # Remove noise from the image
    restored_image = remove_noise(noisy_image, remove_scale)

    # Display the restored image
    plt.figure(figsize=(10, 5))
    plt.imshow(restored_image, cmap='gray')
    plt.title(f'Restored Image (Remove Scale: {remove_scale})')
    plt.axis('off')
    plt.show()

    # Decrement scale by 20 for removing noise
    remove_scale -= 20
```

# **Noise Normal**

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Path to the image in your Google Drive
image_path = 'image.jpg'
# Load the image
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
print("Original image shape:", image.shape)

# Function to add normal noise to the image
def add_normal_noise(image, mean, std_dev):
    # Generate random noise based on normal distribution
    noise = np.random.normal(mean, std_dev, size=image.shape).astype(np.uint8)
    # Apply the noise to the image
    noisy_image = cv2.add(image, noise)
    return noisy_image

# Function to remove noise from the image
def remove_noise(image, mean, std_dev):
    # Estimate the noise based on the normal distribution
    estimated_noise = np.random.normal(mean, std_dev, size=image.shape).astype(np.uint8)
    # Remove the estimated noise from the image
    denoised_image = cv2.subtract(image, estimated_noise)
    return denoised_image

# Start parameters for adding noise
add_mean = 0
add_std_dev = 20

# Loop until std_dev reaches 100 for adding noise
while add_std_dev <= 100:
    # Add normal noise to the image
    noisy_image = add_normal_noise(image, add_mean, add_std_dev)

    # Display the noisy image
    plt.figure(figsize=(10, 5))
    plt.imshow(noisy_image, cmap='gray')
    plt.title(f'Noisy Image (Add Mean: {add_mean}, Add Std Dev: {add_std_dev})')
    plt.axis('off')
    plt.show()

    # Increment std_dev by 20 for adding noise
    add_std_dev += 20

# Start parameters for removing noise
remove_mean = 0
remove_std_dev = 100

# Loop until std_dev reaches 0 for removing noise
while remove_std_dev >= 0:
    # Remove noise from the image
    restored_image = remove_noise(noisy_image, remove_mean, remove_std_dev)

    # Display the restored image
    plt.figure(figsize=(10, 5))
    plt.imshow(restored_image, cmap='gray')
    plt.title(f'Restored Image (Remove Mean: {remove_mean}, Remove Std Dev: {remove_std_dev})')
    plt.axis('off')
    plt.show()

    # Decrement std_dev by 20 for removing noise
    remove_std_dev -= 20
```
